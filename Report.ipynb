```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import os

# Check if the directory exists and list its contents
file_path = '/content/drive/MyDrive/milaireport/APT.csv'
directory = os.path.dirname(file_path)

if not os.path.exists(directory):
    print(f"Error: Directory '{directory}' does not exist. Please check the path.")
else:
    print(f"Contents of '{directory}':")
    for item in os.listdir(directory):
        print(item)

# Attempt to read the CSV file
try:
    df = pd.read_csv(file_path)
    print("File read successfully!")
    print(df.head())
except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found. Please ensure the file exists and the path is correct.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```
